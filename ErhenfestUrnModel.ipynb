```python
# 0 < a, b < 1:

import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

def Path(n, b, a, bull, bear):
    Bull_Path = np.zeros([n])
    Bear_Path = np.zeros([n])
    Bull_Path[0] = bull
    Bear_Path[0] = bear
    total = bear + bull
    for i in range(1,n):
        p1 = (bull/total)*((b + bear)/(a + b + total - 1))
        p2 = (bull/total)*((a + bull - 1)/(a + b + total - 1)) + (bear/total)*((b + bear - 1)/(a + b + total - 1))
        p3 = (bear/total)*((a + bull)/(a + b + total - 1))
        x = np.random.choice(3, p = [p1,p2, p3])
        if x == 0:
            bull = bull - 1
            bear = bear + 1
        if x == 1:
            bull = bull
            bear = bear
        if x == 2:
            bull = bull + 1
            bear = bear - 1
              
        Bull_Path[i] = bull
        Bear_Path[i] = bear
    return (Bull_Path, Bear_Path)

#Case 1
# 0 < a, b < 1:
# Results display long periods of time spent near the boundaries, along with 
# quick fluctuations centered around the origin. 
b = 0.2 
a = 0.8
bull = 50 
bear = 50
n = 1000000

Path = Path(n, b, a, bull, bear)
x = np.arange(0,n)
y1 = Path[0]
y2 = Path[1]

plt.figure(figsize=(14,7))
plt.title("Graph of number of Bullish People at time T")
plt.xlabel('Number of transitions')
plt.ylabel('Number of Bullish People')
plt.plot(x, y1, color="red")

plt.show()
```

```python
#this is the function of the price of the stock over time 
def Stockprice(n, b, a, bull, bear, S0):
    Bull_Path = np.zeros([n])
    Bear_Path = np.zeros([n])
    D_t = np.zeros([n])
    Bull_Path[0] = bull
    Bear_Path[0] = bear
    total = bear + bull
    Stockprice = np.zeros([n])
    Stockprice[0] = S0
    for i in range(1,n):
        p1 = (bull/total)*((b + bear)/(a + b + total - 1))
        p2 = (bull/total)*((a + bull - 1)/(a + b + total - 1)) + (bear/total)*((b + bear - 1)/(a + b + total - 1))
        p3 = (bear/total)*((a + bull)/(a + b + total - 1))
        x = np.random.choice(3, p = [p1,p2, p3])
        if x == 0:
            bull = bull - 1
            bear = bear + 1
        if x == 1:
            bull = bull
            bear = bear
        if x == 2:
            bull = bull + 1
            bear = bear - 1
              
        Bull_Path[i] = bull
        Bear_Path[i] = bear
        D_t[i] = bull - bear
        Stockprice[i] = Stockprice[i-1] + D_t[i]
    return (Stockprice)

bull = 20
bear = 20
b = 10
a = 15
n = 1000
S0 = 1000
Stockprice = Stockprice(n, b, a, bull, bear, S0)
x = np.arange(0,n)
y1 = Stockprice

plt.figure(figsize=(14,7))
plt.title("Graph of Stock price over time ")
plt.xlabel('Number of transitions')
plt.ylabel('Stock Price')
plt.plot(x, y1, color="red")
#plt.plot(x, y2, color="blue")

plt.show()
```

```python
#This is the block which gives us the data we need for 1 simulation
def StockpriceSimulation(b, a, bull, bear, S0, Ts, Tm, Th):
    Bull_Path = np.zeros([Th])
    Bear_Path = np.zeros([Th])
    D_t = np.zeros([Th])
    Bull_Path[0] = bull
    Bear_Path[0] = bear
    total = bear + bull
    Stockprice = np.zeros([Th])
    Stockprice[0] = S0
    data = []
    for i in range(1,Th):
        p1 = (bull/total)*((b + bear)/(a + b + total - 1))
        p2 = (bull/total)*((a + bull - 1)/(a + b + total - 1)) + (bear/total)*((b + bear - 1)/(a + b + total - 1))
        p3 = (bear/total)*((a + bull)/(a + b + total - 1))
        x = np.random.choice(3, p = [p1,p2, p3])
        if x == 0:
            bull = bull - 1
            bear = bear + 1
        if x == 1:
            bull = bull
            bear = bear
        if x == 2:
            bull = bull + 1
            bear = bear - 1
              
        Bull_Path[i] = bull
        Bear_Path[i] = bear
        D_t[i] = bull - bear
        Stockprice[i] = Stockprice[i-1] + D_t[i]
    data.append(Stockprice[Ts-1])  #price at time Ts
    data.append(Stockprice[Tm-1])  #price at time Tm
    data.append(Stockprice[Th-1])  #price at time Th
    data.append((Stockprice[Ts-1]-S0)/S0)  #return at time Ts
    data.append((Stockprice[Tm-1]-Stockprice[Ts-1])/Stockprice[Ts-1])  #return at time Tm
    data.append((Stockprice[Th-1]-Stockprice[Tm-1])/Stockprice[Tm-1])  #return at time TH
    return data #should return an array of 6 pieces of information  
                
                
```
